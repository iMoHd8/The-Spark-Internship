## Removing The Numerical Columns of the columns that has duplicates

```python
df.drop(['country', 'region', 'attacktype1', 'targtype1', 'targsubtype1', 'natlty1', 'weaptype1', 'weapsubtype1'], axis=1, inplace=True)
```

```python
df.head()
```

## we can notice that the imonth and iday columns has some zeros as a values, since it doesn't make since we will remove the zeros values from these two columns

```python
df = df[df['imonth'] != 0]
df = df[df['iday'] != 0]
```

## Creating a new date column that contains the full date format

```python
import datetime
df['date'] = df.apply(lambda x: datetime.date(x['iyear'], x['imonth'], x['iday']), axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

## Reposition a date column to be the first column

```python
column_to_move = df.pop("date")
df.insert(0, "date", column_to_move)
```

```python
df.head()
```

## Renaming some columns (Remove ' _txt')

```python
# Create a dictionary to map old names to new names
new_columns = {col: col.replace('_txt', '') for col in df.columns}
# Rename the columns
df = df.rename(columns=new_columns)
```
